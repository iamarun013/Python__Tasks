### Q3) Difference between jupyter and notepad?
Jupyter notebooks provide an interactive environment where you can execute code in cells and see the output immediately. It allows for a more exploratory and iterative workflow. On the other hand, a text editor like Notepad only provides a plain text editing environment and does not have built-in interactivity.

### Q4) Write Down difference between interpreter and compiler
Intrepreter
Translates program one statement at a time.

Interpreters usually take less amount of time to analyze the source code.

Programming languages like JavaScript, Python, Ruby use interpreters.

Compiler
Scans the entire program and translates it as a whole into machine code.

Compilers usually take a large amount of time to analyze the source code.

Programming languages like C, C++, Java use compilers.
